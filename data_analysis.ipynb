```python
# 假设输入是 DataFrame
def check_sp_overlap_by_BG(df):
    """
    检查按照 BG 分组后，组装厂是否存在重叠。

    :param df: 输入的 DataFrame, 包含 BG 和组装厂列
    """
    # 按照 BG 分组
    subtables = df.groupby(WeekCapacity.BG)
    
    # 提取每个子表的组装厂集合
    sp_sets = {}
    for bg, group in subtables:
        sp_set = set(group[WeekCapacity.FACTORY])  # 只看组装厂
        sp_sets[bg] = sp_set
    
    # 检查子表之间的组装厂是否存在重叠
    bgs = list(sp_sets.keys())
    has_overlap = False
    
    if len(bgs) < 2:
        raise ValueError("仅存在一个 BG 类别")
    
    for i in range(len(bgs)):
        for j in range(i + 1, len(bgs)):
            bg1, bg2 = bgs[i], bgs[j]
            common_sp = sp_sets[bg1].intersection(sp_sets[bg2])
            if common_sp:
                has_overlap = True
                print(f"BG {bg1} 和 BG {bg2} 之间存在重叠的组装厂: {common_sp}")
    
    # 输出结果
    if not has_overlap:
        print("按照 BG 划分后, 不存在组装厂的重叠。")
    else:
        print("按照 BG 划分后, 仍然存在组装厂的重叠。")


check_sp_overlap_by_BG(Z_capacity)
```

```python
check_sp_overlap_by_BG(Y_capacity)
```

```python
# 假设输入是DataFrame
def check_sp_pl_overlap_by_PDT(df):
    # 按照BG分组
    subtables = df.groupby(WeekCapacity.PDT)
    
    # 提取每个子表的(组装厂, 产线)对
    sp_pl_pairs = {}
    for pdt, group in subtables:
        pairs = set(group[WeekCapacity.SP_PL_PAIR])
        sp_pl_pairs[pdt] = pairs
    
    # 检查子表之间的(组装厂, 产线)对是否存在重叠
    pdts = list(sp_pl_pairs.keys())
    has_overlap = False
    
    if len(pdts) < 2:
        raise ValueError("仅存在一个 PDT 类别")
    
    for i in range(len(pdts)):
        for j in range(i + 1, len(pdts)):
            pdt1, pdt2 = pdts[i], pdts[j]
            common_pairs = sp_pl_pairs[pdt1].intersection(sp_pl_pairs[pdt2])
            if common_pairs:
                has_overlap = True
                print(f"PDT {pdt1} 和 PDT {pdt2} 之间存在重叠的(组装厂, 产线)对: {common_pairs}")
    
    # 输出结果
    if not has_overlap:
        print("按照PDT划分后, 不存在(组装厂, 产线)对的重叠。")
    else:
        print("按照PDT划分后, 仍然存在(组装厂, 产线)对的重叠。")

# 示例调用（假设df是你的DataFrame）
check_sp_pl_overlap_by_PDT(Z_capacity)
```

```python
check_sp_pl_overlap_by_PDT(Y_capacity)
```
